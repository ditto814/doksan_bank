```python
# 라이브러리 임포트

from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
from selenium.webdriver.common.by import By
import codecs
```

```python
# URL에 접근하여 소스 저장
driver = webdriver.Chrome()
time.sleep(2)

driver.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=2&cnt=1000&SortOrder=1')
time.sleep(2)

source = driver.page_source

print(type(source))
print(source)
```

```python
# 소스 파싱(페이지 전체)

parsed_source = BeautifulSoup(source, 'html.parser')
parsed_source
```

```python
# li 태그 갯수
len(parsed_source.find_all('li'))
```

```python
# 한 페이지에 있는 책만 가져오기(50위씩 도서 페이지 구분)

div_book_list = parsed_source.find_all('div', class_ = 'ss_book_box')
len(div_book_list)
```

```python
# bs4.resultest 자료형은 find_all을 받을 수 없습니다.(리스트로 취급)
# 따라서, 내부 요소 하나하나를 개별적으로 처리해야합니다.
# 1.인덱싱작업 수행
# 2.반복문 내부에서 하나씩 돌아가며 대입하여 처리

# 1위책에 대한 li를 전부 가져옵니다.
# div_book_list[0].find_all('li')
book_li_list = div_book_list[0].find_all('li')
```

```python
# 직접 하나씩 li 인덱스 확인

# print(div_book_list[0].find_all('li')[0].text)
# print(div_book_list[0].find_all('li')[1].text)
# print(div_book_list[0].find_all('li')[2].text)
# print(div_book_list[0].find_all('li')[3].text)
# print(div_book_list[0].find_all('li')[4].text)
# print(div_book_list[0].find_all('li')[5].text)
# print(div_book_list[0].find_all('li')[6].text)
# print(div_book_list[0].find_all('li')[7].text)

print(book_li_list[0].text)
print(book_li_list[1].text)
print(book_li_list[2].text)
print(book_li_list[3].text)
print(book_li_list[4].text)
print(book_li_list[5].text)
print(book_li_list[6].text)
print(book_li_list[7].text)
```

```python
# 하나씩 말고 for문으로 알아보는 방법
for i in range(len(book_li_list)):
    print(book_li_list[i].text)
```

```python
# 50개짜리를 div 
for book in div_book_list:
    # li만 다시 조회해서 변수에 저장
    book_li_list = book.find_all('li')

    # <li>에 증정품 정보가 없는 책은 어떻게 접근하면 될까요?
    # 1번째 요소가 경품이 있으면 책 제목, 경품이 없으면 저자임을 이용하여 
    # 책 제목의 0번째 글자는 무조건 대괄호임을 이용해봅시다.

    if(book_li_list[1].text[0] == '['):
        print(book_li_list[1].text)
        print(book_li_list[2].text)
        print(book_li_list[3].text)
    
    else:
        print(book_li_list[0].text)
        print(book_li_list[1].text)
        print(book_li_list[2].text)

    print('-' * 50)
        
```

```python
# csv에 보기 좋은 방식
for book in div_book_list:

    book_li_list = book.find_all('li')



    if(book_li_list[1].text[0] == '['):
        print(book_li_list[1].text.split('[국내도서]')[1],'&',
              book_li_list[2].text.split('|')[0],'&',
              book_li_list[3].text.split(' ')[2].replace(',',''))
    
    else:
        print(book_li_list[0].text.split('[국내도서]')[1],'&',
              book_li_list[1].text.split('|')[0],'&',
              book_li_list[2].text.split(' ')[2].replace(',',''))
```

```python
f = codecs.open("C:/doksan_bank/02.crawler/aladin__result.txt"  , mode = 'w', encoding = 'utf-8')

for book in div_book_list:

    book_li_list = book.find_all('li')

    if(book_li_list[1].text[0] == '['):

        f.write(book_li_list[1].text.split('[국내도서]')[1])
        f.write(book_li_list[2].text.split('|')[0])
        f.write(book_li_list[3].text.split(' ')[2].replace(',',''))
        f.write(('\n'))
    
    else:
        f.write(book_li_list[0].text.split('[국내도서]')[1])
        f.write(book_li_list[1].text.split('|')[0])
        f.write(book_li_list[2].text.split(' ')[2].replace(',',''))
        f.write(('\n'))
```
