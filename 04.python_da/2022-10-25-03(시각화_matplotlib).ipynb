### Matplotlib
- 파이썬 기반 데이터 분석 환경에서 가장 일반적으로 사용되는 시각화 라이브러리
- 하위 라이브러리인 'pyplot'을 통해 주요 기능을 사용
- plotly, seaborn 등과 같은 라이브러리가 최근 경쟁력을 갖추고 있음
- pandas 객체의 기본적인 시각화 도구
- plot() 메서드를 통해 데이터를 시각화
- 종류
    - 라인 플롯(line plot)
    - 바 차트(bar chart)
    - 히스토그램(histogram)
    - 박스 플롯(box plot)
    - 스캐터 플롯(scatter plot)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### plt.plot(y)
- 기본 : 라인 플롯(line plot)
- 데이터가 시간, 순서 등에 따라 변화하는 흐름 파악
- x가 주어지지 않을 경우 자동 생성 :x = range(len(y))

```python
# 데이터 생성 : 5~6개의 아이템을 갖는 리스트
# plot(data) : 기본적으로 y축의 값으로 데이터를 매칭
# x -> len(data) 만킁의 인덱스번호를 생성
# 기본 plot = line plot
plt.plot([1,3,4,2,6,10])
plt.show()
```

```python
# x tick(x축 값) 설정
# plt.plot(x, y) : x에 넣어준 리스트와 y에 엏어준 리스트의 길이는 같아야합니다.
plt.plot([2,4,6,8,10,12], [1,3,4,2,6,10])
plt.show()
```

### 스타일 지정
- plt.plot([x], y, [fmt], **kwargs)
    - fmt : '[color][marker][line]'
        - 색상, 마커, 선 등에 대한 포맷 옵션 설정
1) 마커 참조 도큐먼트 주소<br>
https://matplotlib.org/3.1.3/api/markers_api.html<br>
2) 색상 참조 도큐먼트 주소<br>
https://matplotlib.org/3.1.0/gallery/color/named_colors.html<br>
3) 라인 스타일 참조 도큐먼트 주소<br>
https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html<br>

```python
# 스타일 문자열 추가 : [fmt] -> color + marker + line 순서대로 문자열 작성
# 어느 요소를 생략하면 해당 생략된 요소는 기본값으로 적용
# 'o' : 라린 없이 큰 점(원모양)으로 데이터 위치 표시
plt.plot([1,2,3,4], [2,6,10,7], 'o')
plt.show
```

```python
# o- : 마커는 동그라미, 라인은 실선
# fmt[색상][마커][선종류]
plt.plot([1,2,3,4], [2,6,10,7], 'o-')
plt.show
```

```python
# o-- : 마커는 동그라미, 라인은 숨은선
# fmt[색상][마커][선종류]
plt.plot([1,2,3,4], [2,6,10,7], 'o--')
plt.show
```

```python
# mo-- : 마커는 동그라미, 라인은 숨은선, 색은 마젠타색
# fmt[색상][마커][선종류]
plt.plot([1,2,3,4], [2,6,10,7], 'mo--')
plt.show
```

```python
# 기타 스타일 파라미터 종류
# color : 색상(magenta)
# linestyle : -.
# marker : ^
# markerfacecolor : orange
# markersize : 20
plt.plot([1, 2, 3, 4], [2, 6, 10, 7],
        color = 'magenta', linestyle='-.', marker='^',
        markeredgecolor='r', markerfacecolor='orange', markersize='20')
plt.show()
```

### 축 범위 지정
#### matplotlib.pyplot.axis
- xmin, xmax, ymin, ymax = plt.axis([xmin, xmax, ymin, ymax])
    - xmin : x축 최소값
    - xmax : x축 최대값
    - ymin : y축 최소값
    - ymax : y축 최대값
#### matplotlib.pyplot.xlim / ylim
- plt.xlim(min, max)
- plt.ylim(min, max)

```python
plt.plot([1,2,3,4], [10,3,7,13])
plt.show()
```

```python
plt.plot([1,2,3,4], [10,3,7,13])
# 그래프가있는 백그라운드 크기 조절
plt.axis([0,7, 2, 20])
plt.show()
```

```python
# 선에 비해 지나치게 작은 플롯을 구성하면 일부 플롯이 잘려나갑니다.
plt.plot([1,2,3,4], [10,3,7,13])
plt.axis([2, 4, 4, 10])
plt.show()
```

```python
# grid 표시하기
# grid는 격자무늬로 표시할 때는 plt.grid(True)를 한 줄 더 실행합니다.
plt.plot([1,2,3,4], [10,3,7,13])
plt.grid(True)
plt.show()
```

#### matplotlib.pyplot.scatter
- 대표값을 중심으로 데이터가 흩어져 있는 정도를 시각화
- 데이터의 분산 정도, 데이터 간의 상관관계를 파악할 수 있음
- plt.scatter(x, y, s=None, c=None, data=None)
    - s : 마커 사이즈
    - c : 마커 색상
    - data : dictionary 자료형을 사용하여 데이터를 전달하면 keyword string을 통해 데이터 사용 가능

```python
np.random.seed(2022)
```

```python
# 정규분포 확률밀도에서 100개의 표본추출
# rando.normal
x = np.random.normal(0,1,100)
y = np.random.normal(0,1,100)
plt.scatter(x,y)
plt.show()
```

```python
print(x[0])
print(y[0])
```

```python
# plot으로 하면 상당히 분간하기 어렵기 때문에 산점도는 scatter를 주로 사용합니다.
plt.plot(x,y)
plt.show()
```

```python
# c는 산점도 색, s는 범례 사이즈
plt.scatter(x,y, c='gold', s = 100)
plt.show()
```

### Bar chart (막대 그래프)
- 여러 값의 상대적인 크기를 비교하기에 적합한 표현법이며 원형 차트보다 더욱 일반적으로 사용됨
- plt.bar(x, y)
    - x 데이터가 카테고리 값인 경우에는 bar, barh 로 바 차트 시각화 표현
    - 가로 방향으로 바 차트를 그리려면 barh
    - barh 차트 주의점 : x축이 일반적으로 y축의 위치를 의미

```python
# 범주형 데이터에 대한 bar graph 기본
x = ['가', '니', '다']
y = [2, 10, 5]
plt.bar(x,y)
plt.show()
```

```python
# 제목, xlabel, ylabel
# 막대마다 붙일 이름을 바꿔주고 싶다면, xticks를 사용합니다.
plt.bar(x,y)
plt.title('bar chart')
plt.xlabel('x축') 
plt.ylabel('y축')
plt.xticks(x, ['김', '이', '박'])

plt.show()
```

```python
# 각각의 bar마다 색상 변경
# plt.bar()객체 set_color()
bar_list = plt.bar(x,y)
bar_list[0].set_color('r')
bar_list[1].set_color('g')
plt.title('bar chart')
plt.xticks(x, ['김', '이', '박'])
plt.xlabel('x축') 
plt.ylabel('y축')
plt.show()
```
