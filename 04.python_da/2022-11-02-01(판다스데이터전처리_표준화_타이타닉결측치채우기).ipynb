##### 남성은 0, 여성은 1로 가지는 sex_num 컬럼 추가
    - 남성 : male , 여성 : female
    - 남성 : 0, 여성 : 1
    - 문자타입 범주형 -> 숫자타입 범주형

```python
# 방법 1 : {'male' : 0, 'female':1} 로 접근하는 방법
# 방법 2 : def로 만들어서 접근하는 방법 (내가 선택한 방법)
data
```

```python
# 성별마다 번호를 할당하는 함수 작성
def gender_num(value):
    if value == 'female':
        return 1
    else:
        return 0

# 성별마다 번호를 할당하는 함수 적용
data['Sex_num'] = data['Sex'].apply(gender_num)
# 다른 작성법
# data['Sex_num'] = data['Sex'].map(gender_num)
data

# tip : 파이썬은 일급객체득성이 있어 변수에 함수를 저장할 수있습니다.(자세한 사항은 일급객체 검색)
```

##### 이름 길이 값을 저장하는 name_length 컬럼추가

```python
# 방법 2 
# 왕 간단한 방법
data['Name_length'] = data['Name'].map(len)
data
```

```python
# 성별이 생존에 영향을 미쳤는지 여부 확인
# 상관계수를 구하면 컬럼간 관계를 유추할 수 있습니다.
# 대개 0.3미만은 무시하는데 0.5 정도면 나름 타당하다는 판단을 할 수 있습니다.
data[['Survived', 'Sex_num']].corr()
```

```python
# 이름길이가 영향을 미쳤는지 여부 확인
# 이름 긴사람들은 부유층이거나 지도층이다 -> 지도층은 생존률이 높다 -> 봤는데 그리크지 않았다 -> 가설 기각 
data[['Survived', 'Name_length']].corr()
```
